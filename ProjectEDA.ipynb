```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/vehicle_sales_dirty.csv")
df.head()
df.info()
```

# Filling the Null Values

```python
for col in ['Brand', 'Fuel_Type', 'Transmission', 'Owner']:
    mode_value = df[col].mode()[0]
    df[col] = df[col].fillna(mode_value)
median_price = df['Selling_Price'].median()
df['Selling_Price'].fillna(median_price, inplace=True)
df['Selling_Price_dup'].fillna(median_price, inplace=True)
df.isnull().sum()
```

```python
df.to_csv("cleaned_vehicle_sales.csv", index=False)
```

# Cleaned Data

```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/cleaned_vehicle_sales.csv")
df.head()
df.info()
```

# Categorization of **Data**

```python
numerical_cols = df.select_dtypes(include=np.number)
categorical_cols = df.select_dtypes(include='object')

print("Numerical Columns:")
display(numerical_cols.head())

print("\nCategorical Columns:")
display(categorical_cols.head())
```

# Heat Map

```python
import seaborn as sns
import matplotlib.pyplot as plt
correlation_matrix = numerical_cols.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```

# Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.scatter(df['Selling_Price'],df['Driven_KMs'])
plt.title('Scatter Plot Matrix of Numerical Columns')
plt.xlabel('Selling_Price')
plt.ylabel('Driven_KMs')
plt.show()
```

# Pie Chart

```python
fuel_counts = df["Fuel_Type"].value_counts()
plt.pie(fuel_counts, labels=fuel_counts.index, autopct="%1.0f%%")
plt.title("Pie Chart - Fuel Type Distribution")
plt.show()
```

# Bar Chart

```python
plt.figure(figsize=(8,5))
df.groupby('Brand')['Driven_KMs'].mean().plot(kind='bar')
plt.bar(df['Brand'],df['Driven_KMs'])
plt.title('Brand vs distances')
plt.xlabel('Category')
plt.ylabel('distance')
plt.show()
```
